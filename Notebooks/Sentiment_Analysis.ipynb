```python
# --- Convert timestamps and align datasets by date (daily level) ---

# 1. Fear/Greed: convert timestamp (Unix seconds) and ensure date is datetime
fear_df['datetime'] = pd.to_datetime(fear_df['timestamp'], unit='s')
fear_df['date'] = pd.to_datetime(fear_df['date']).dt.normalize()

# 2. Historical trades: parse Timestamp IST (DD-MM-YYYY HH:MM) and Timestamp (Unix ms)
hist_df['datetime_ist'] = pd.to_datetime(hist_df['Timestamp IST'], format='%d-%m-%Y %H:%M')
# Timestamp column: if > 1e12 assume milliseconds, else seconds
hist_df['datetime_ts'] = hist_df['Timestamp'].apply(
    lambda x: pd.to_datetime(x, unit='ms') if x >= 1e12 else pd.to_datetime(x, unit='s')
)
hist_df['date'] = hist_df['datetime_ist'].dt.normalize()
```

```python
# 3. Aggregate historical trades to daily level
hist_daily = hist_df.groupby('date').agg(
    trade_count=('Trade ID', 'nunique'),
    total_volume_usd=('Size USD', 'sum'),
    total_pnl=('Closed PnL', 'sum'),
    total_fees=('Fee', 'sum'),
).reset_index()
buy_vol = hist_df[hist_df['Side'] == 'BUY'].groupby('date')['Size USD'].sum().reset_index()
buy_vol.columns = ['date', 'buy_volume_usd']
hist_daily = hist_daily.merge(buy_vol, on='date', how='left')
hist_daily['buy_volume_usd'] = hist_daily['buy_volume_usd'].fillna(0)
```

```python
# 4. Align datasets: merge Fear/Greed with daily trading metrics on date
aligned_df = hist_daily.merge(
    fear_df[['date', 'value', 'classification']].drop_duplicates('date'),
    on='date',
    how='inner'
)
aligned_df = aligned_df.sort_values('date').reset_index(drop=True)
print(f"Aligned dataset: {len(aligned_df)} days | Date range: {aligned_df['date'].min().date()} to {aligned_df['date'].max().date()}")
aligned_df.head(10)
```

```python
# --- Key metrics for analysis (per-trade -> daily per account -> daily overall) ---

# Normalize side
hist_df['side'] = hist_df['Side'].astype(str).str.upper().str.strip()
hist_df['is_buy'] = hist_df['side'].eq('BUY').astype(int)
hist_df['is_sell'] = hist_df['side'].eq('SELL').astype(int)

# Trade economics
hist_df['fee_usd'] = hist_df['Fee'].astype(float)
hist_df['closed_pnl_usd'] = hist_df['Closed PnL'].astype(float)
hist_df['net_pnl_usd'] = hist_df['closed_pnl_usd'] - hist_df['fee_usd']

# Win/loss: only evaluate on rows where PnL is realized (Closed PnL != 0)
hist_df['is_pnl_event'] = hist_df['closed_pnl_usd'].ne(0)
hist_df['is_win'] = (hist_df['net_pnl_usd'] > 0) & hist_df['is_pnl_event']

# Trade sizing
hist_df['trade_size_usd'] = hist_df['Size USD'].astype(float)
hist_df['trade_size_tokens'] = hist_df['Size Tokens'].astype(float)

# Buy/Sell volume (proxy for long/short pressure)
hist_df['buy_volume_usd'] = np.where(hist_df['is_buy'].eq(1), hist_df['trade_size_usd'], 0.0)
hist_df['sell_volume_usd'] = np.where(hist_df['is_sell'].eq(1), hist_df['trade_size_usd'], 0.0)

# Position size proxy (true leverage isn't available in this export)
# Start Position appears to be in tokens; convert to USD notional using execution price.
hist_df['start_pos_tokens'] = hist_df['Start Position'].astype(float)
hist_df['execution_price'] = hist_df['Execution Price'].astype(float)
hist_df['start_pos_notional_usd'] = (hist_df['start_pos_tokens'].abs() * hist_df['execution_price']).astype(float)

hist_df[['date','Account','Coin','side','trade_size_usd','closed_pnl_usd','fee_usd','net_pnl_usd','start_pos_notional_usd']].head()
```

```python
# Daily metrics per account (trader behavior + performance)
acct_daily = hist_df.groupby(['date', 'Account']).agg(
    # activity
    fills=('trade_size_usd', 'size'),
    trade_count=('Trade ID', 'nunique'),
    buy_fills=('is_buy', 'sum'),
    sell_fills=('is_sell', 'sum'),

    # sizing
    total_volume_usd=('trade_size_usd', 'sum'),
    avg_trade_size_usd=('trade_size_usd', 'mean'),
    median_trade_size_usd=('trade_size_usd', 'median'),
    p90_trade_size_usd=('trade_size_usd', lambda s: s.quantile(0.90)),
    p95_trade_size_usd=('trade_size_usd', lambda s: s.quantile(0.95)),

    # performance
    total_closed_pnl_usd=('closed_pnl_usd', 'sum'),
    total_fees_usd=('fee_usd', 'sum'),
    total_net_pnl_usd=('net_pnl_usd', 'sum'),
    pnl_event_count=('is_pnl_event', 'sum'),
    win_count=('is_win', 'sum'),

    # long/short proxy (volume)
    buy_volume_usd=('buy_volume_usd', 'sum'),
    sell_volume_usd=('sell_volume_usd', 'sum'),

    # position-size proxy distribution (true leverage isn't in this export)
    max_start_pos_notional_usd=('start_pos_notional_usd', 'max'),
    mean_start_pos_notional_usd=('start_pos_notional_usd', 'mean'),
).reset_index()

# derived ratios
acct_daily['win_rate'] = np.where(
    acct_daily['pnl_event_count'] > 0,
    acct_daily['win_count'] / acct_daily['pnl_event_count'],
    np.nan,
)
acct_daily['net_flow_usd'] = acct_daily['buy_volume_usd'] - acct_daily['sell_volume_usd']
acct_daily['buy_volume_share'] = acct_daily['buy_volume_usd'] / (acct_daily['total_volume_usd'] + 1e-9)
acct_daily['buy_fill_share'] = acct_daily['buy_fills'] / (acct_daily['fills'] + 1e-9)

# keep ratio, but avoid exploding values when sell_volume_usd == 0
acct_daily['buy_sell_ratio'] = np.where(
    acct_daily['sell_volume_usd'] > 0,
    acct_daily['buy_volume_usd'] / acct_daily['sell_volume_usd'],
    np.nan,
)

acct_daily.head()
```

```python
# Daily overall metrics (market-wide behavior + performance)
daily_metrics = acct_daily.groupby('date').agg(
    active_traders=('Account', 'nunique'),

    # activity
    trades=('trade_count', 'sum'),
    fills=('fills', 'sum'),
    buy_fills=('buy_fills', 'sum'),
    sell_fills=('sell_fills', 'sum'),

    # sizing
    total_volume_usd=('total_volume_usd', 'sum'),
    avg_trade_size_usd=('avg_trade_size_usd', 'mean'),

    # performance
    total_closed_pnl_usd=('total_closed_pnl_usd', 'sum'),
    total_fees_usd=('total_fees_usd', 'sum'),
    total_net_pnl_usd=('total_net_pnl_usd', 'sum'),
    pnl_event_count=('pnl_event_count', 'sum'),
    win_count=('win_count', 'sum'),

    # long/short proxy (volume)
    buy_volume_usd=('buy_volume_usd', 'sum'),
    sell_volume_usd=('sell_volume_usd', 'sum'),

    # position-size proxy distribution across traders
    p50_max_start_pos_notional_usd=('max_start_pos_notional_usd', lambda s: s.quantile(0.50)),
    p75_max_start_pos_notional_usd=('max_start_pos_notional_usd', lambda s: s.quantile(0.75)),
    p90_max_start_pos_notional_usd=('max_start_pos_notional_usd', lambda s: s.quantile(0.90)),
    p95_max_start_pos_notional_usd=('max_start_pos_notional_usd', lambda s: s.quantile(0.95)),
    p99_max_start_pos_notional_usd=('max_start_pos_notional_usd', lambda s: s.quantile(0.99)),
).reset_index()

# derived daily ratios
_daily_traders = daily_metrics['active_traders'] + 1e-9
_daily_volume = daily_metrics['total_volume_usd'] + 1e-9

daily_metrics['pnl_per_trader'] = daily_metrics['total_net_pnl_usd'] / _daily_traders

daily_metrics['win_rate'] = np.where(
    daily_metrics['pnl_event_count'] > 0,
    daily_metrics['win_count'] / daily_metrics['pnl_event_count'],
    np.nan,
)

daily_metrics['net_flow_usd'] = daily_metrics['buy_volume_usd'] - daily_metrics['sell_volume_usd']
daily_metrics['buy_volume_share'] = daily_metrics['buy_volume_usd'] / _daily_volume

daily_metrics['buy_fill_share'] = daily_metrics['buy_fills'] / (daily_metrics['fills'] + 1e-9)

# keep ratio, but avoid exploding values when sell_volume_usd == 0
daily_metrics['buy_sell_ratio'] = np.where(
    daily_metrics['sell_volume_usd'] > 0,
    daily_metrics['buy_volume_usd'] / daily_metrics['sell_volume_usd'],
    np.nan,
)

# Align with Fear/Greed at daily level
aligned_daily = daily_metrics.merge(
    fear_df[['date', 'value', 'classification']].drop_duplicates('date'),
    on='date',
    how='inner'
).sort_values('date').reset_index(drop=True)

aligned_daily[[
    'date','value','classification',
    'active_traders','trades','fills','total_volume_usd',
    'total_net_pnl_usd','pnl_per_trader','win_rate',
    'buy_volume_share','buy_fill_share','net_flow_usd','buy_sell_ratio',
    'p95_max_start_pos_notional_usd'
]].head(10)
```

```python
# --- Evidence: Fear vs Greed impact on performance & behavior ---

# Sentiment regime: collapse into Fear / Greed / Neutral
fear_map = {
    'Extreme Fear': 'Fear',
    'Fear': 'Fear',
    'Neutral': 'Neutral',
    'Greed': 'Greed',
    'Extreme Greed': 'Greed',
}

fear_df['regime'] = fear_df['classification'].map(fear_map)

# Align account-daily metrics with sentiment
aligned_acct_daily = acct_daily.merge(
    fear_df[['date', 'value', 'classification', 'regime']].drop_duplicates('date'),
    on='date',
    how='inner'
).sort_values(['date', 'Account']).reset_index(drop=True)

# Keep only Fear vs Greed for comparisons
fg_daily = aligned_daily[aligned_daily['classification'].isin(['Extreme Fear','Fear','Greed','Extreme Greed'])].copy()
fg_daily['regime'] = fg_daily['classification'].map(fear_map)

fg_acct = aligned_acct_daily[aligned_acct_daily['regime'].isin(['Fear','Greed'])].copy()

print('Days (Fear vs Greed):', fg_daily['date'].nunique())
print('Acct-days (Fear vs Greed):', len(fg_acct))
fg_daily[['date','classification','regime']].head()
```

```python
# 1) Performance differs between Fear vs Greed days (PnL, win rate, drawdown proxy)

# Drawdown proxy (per account): compute cumulative net PnL and drawdown over time
fg_acct = fg_acct.sort_values(['Account','date'])
fg_acct['cum_net_pnl'] = fg_acct.groupby('Account')['total_net_pnl_usd'].cumsum()
fg_acct['cum_peak'] = fg_acct.groupby('Account')['cum_net_pnl'].cummax()
fg_acct['drawdown_proxy'] = fg_acct['cum_net_pnl'] - fg_acct['cum_peak']  # <= 0

# Summaries by regime (account-day level)
perf_by_regime = fg_acct.groupby('regime').agg(
    acct_days=('Account', 'size'),
    traders=('Account', 'nunique'),
    net_pnl_mean=('total_net_pnl_usd', 'mean'),
    net_pnl_median=('total_net_pnl_usd', 'median'),
    win_rate_mean=('win_rate', 'mean'),
    win_rate_median=('win_rate', 'median'),
    drawdown_mean=('drawdown_proxy', 'mean'),
    drawdown_p05=('drawdown_proxy', lambda s: s.quantile(0.05)),
).reset_index()

perf_by_regime
```

```python
# Same comparison at day level (market-wide)
daily_by_regime = fg_daily.groupby('regime').agg(
    days=('date','nunique'),
    active_traders_mean=('active_traders','mean'),
    trades_mean=('trades','mean'),
    fills_mean=('fills','mean'),
    total_volume_usd_mean=('total_volume_usd','mean'),
    pnl_per_trader_mean=('pnl_per_trader','mean'),
    pnl_per_trader_median=('pnl_per_trader','median'),
    win_rate_mean=('win_rate','mean'),
    buy_volume_share_mean=('buy_volume_share','mean'),
    net_flow_usd_mean=('net_flow_usd','mean'),
    p95_max_pos_mean=('p95_max_start_pos_notional_usd','mean'),
).reset_index()

daily_by_regime
```

```python
# 2) Behavior changes by sentiment (frequency, size proxy, long/short proxy)
behavior_by_regime = fg_acct.groupby('regime').agg(
    acct_days=('Account','size'),
    fills_mean=('fills','mean'),
    trades_mean=('trade_count','mean'),
    volume_mean=('total_volume_usd','mean'),
    avg_trade_size_mean=('avg_trade_size_usd','mean'),
    buy_volume_share_mean=('buy_volume_share','mean'),
    buy_fill_share_mean=('buy_fill_share','mean'),
    net_flow_usd_mean=('net_flow_usd','mean'),
    max_pos_p50=('max_start_pos_notional_usd', lambda s: s.quantile(0.50)),
    max_pos_p90=('max_start_pos_notional_usd', lambda s: s.quantile(0.90)),
).reset_index()

behavior_by_regime
```

```python
# 3) Trader segments and how they differ (2–3 segments)

# Build per-account summary across all aligned days (Fear+Greed only for clean comparison)
acct_summary = fg_acct.groupby('Account').agg(
    active_days=('date','nunique'),
    total_net_pnl=('total_net_pnl_usd','sum'),
    avg_daily_pnl=('total_net_pnl_usd','mean'),
    pnl_std=('total_net_pnl_usd','std'),
    avg_trades_per_day=('trade_count','mean'),
    avg_fills_per_day=('fills','mean'),
    median_max_pos=('max_start_pos_notional_usd','median'),
    p95_max_pos=('max_start_pos_notional_usd', lambda s: s.quantile(0.95)),
    win_rate_mean=('win_rate','mean'),
).reset_index()

# Drawdown proxy per account (most negative drawdown observed)
acct_dd = fg_acct.groupby('Account')['drawdown_proxy'].min().reset_index().rename(columns={'drawdown_proxy':'max_drawdown_proxy'})
acct_summary = acct_summary.merge(acct_dd, on='Account', how='left')

# Segment A: High vs Low "leverage" proxy using median max position notional
q25 = acct_summary['median_max_pos'].quantile(0.25)
q75 = acct_summary['median_max_pos'].quantile(0.75)
acct_summary['pos_size_segment'] = np.where(
    acct_summary['median_max_pos'] >= q75, 'High position size',
    np.where(acct_summary['median_max_pos'] <= q25, 'Low position size', 'Mid')
)

# Segment B: Frequent vs infrequent traders using avg trades/day
t25 = acct_summary['avg_trades_per_day'].quantile(0.25)
t75 = acct_summary['avg_trades_per_day'].quantile(0.75)
acct_summary['frequency_segment'] = np.where(
    acct_summary['avg_trades_per_day'] >= t75, 'Frequent',
    np.where(acct_summary['avg_trades_per_day'] <= t25, 'Infrequent', 'Mid')
)

# Segment C: Consistency - profitable + lower volatility vs inconsistent
# (heuristic) Consistent winners: total_net_pnl > 0, win_rate_mean > median, pnl_std below median
wr_med = acct_summary['win_rate_mean'].median(skipna=True)
std_med = acct_summary['pnl_std'].median(skipna=True)
acct_summary['consistency_segment'] = np.where(
    (acct_summary['total_net_pnl'] > 0) & (acct_summary['win_rate_mean'] >= wr_med) & (acct_summary['pnl_std'] <= std_med),
    'Consistent winner',
    np.where(acct_summary['total_net_pnl'] > 0, 'Profitable but inconsistent', 'Unprofitable')
)

acct_summary[['Account','active_days','total_net_pnl','avg_trades_per_day','median_max_pos','win_rate_mean','pnl_std','max_drawdown_proxy','pos_size_segment','frequency_segment','consistency_segment']].head()
```

```python
# Segment comparisons (tables)

segA = acct_summary[acct_summary['pos_size_segment'].isin(['Low position size','High position size'])].groupby('pos_size_segment').agg(
    traders=('Account','nunique'),
    total_net_pnl_mean=('total_net_pnl','mean'),
    win_rate_mean=('win_rate_mean','mean'),
    pnl_std_mean=('pnl_std','mean'),
    max_dd_mean=('max_drawdown_proxy','mean'),
    avg_trades_per_day=('avg_trades_per_day','mean'),
    median_max_pos=('median_max_pos','median'),
).reset_index()

segB = acct_summary[acct_summary['frequency_segment'].isin(['Infrequent','Frequent'])].groupby('frequency_segment').agg(
    traders=('Account','nunique'),
    total_net_pnl_mean=('total_net_pnl','mean'),
    win_rate_mean=('win_rate_mean','mean'),
    pnl_std_mean=('pnl_std','mean'),
    max_dd_mean=('max_drawdown_proxy','mean'),
    avg_trades_per_day=('avg_trades_per_day','mean'),
    median_max_pos=('median_max_pos','median'),
).reset_index()

segC = acct_summary.groupby('consistency_segment').agg(
    traders=('Account','nunique'),
    total_net_pnl_mean=('total_net_pnl','mean'),
    win_rate_mean=('win_rate_mean','mean'),
    pnl_std_mean=('pnl_std','mean'),
    max_dd_mean=('max_drawdown_proxy','mean'),
).reset_index().sort_values('traders', ascending=False)

```

```python
# --- Advanced plots (report-quality) ---
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

ROOT = Path.cwd()
if not (ROOT / 'data').exists():
    ROOT = ROOT.parent
REPORTS = ROOT / 'reports'
REPORTS.mkdir(exist_ok=True)

palette = {'Fear': '#2563EB', 'Greed': '#F97316', 'Neutral': '#64748B'}

def _money(x, _pos=None):
    try:
        x = float(x)
    except Exception:
        return str(x)
    ax = abs(x)
    if ax >= 1_000_000_000:
        return f"{x/1_000_000_000:.1f}B"
    if ax >= 1_000_000:
        return f"{x/1_000_000:.1f}M"
    if ax >= 1_000:
        return f"{x/1_000:.1f}k"
    return f"{x:.0f}"

money_fmt = FuncFormatter(_money)

sns.set_theme(
    context='talk',
    style='white',
    font='DejaVu Sans',
    rc={
        'axes.titleweight': 'bold',
        'axes.labelcolor': '#0f172a',
        'text.color': '#0f172a',
        'axes.edgecolor': '#cbd5e1',
        'grid.color': '#e2e8f0',
        'axes.facecolor': 'white',
        'figure.facecolor': 'white',
    },
)

print('Saving plots to', REPORTS)
```

```python
# Chart E: Spearman correlation heatmap (daily metrics + sentiment)

corr_cols = [
    'value',
    'pnl_per_trader',
    'win_rate',
    'trades',
    'fills',
    'total_volume_usd',
    'buy_volume_share',
    'net_flow_usd',
    'p95_max_start_pos_notional_usd',
    'active_traders',
    'avg_trade_size_usd',
]

corr_df = aligned_daily.dropna(subset=['value']).copy()
corr_df = corr_df[[c for c in corr_cols if c in corr_df.columns]]

corr = corr_df.corr(method='spearman')
mask = np.triu(np.ones_like(corr, dtype=bool))

fig, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(
    corr,
    mask=mask,
    cmap=sns.diverging_palette(220, 20, as_cmap=True),
    vmin=-1,
    vmax=1,
    center=0,
    square=True,
    linewidths=0.6,
    cbar_kws={'shrink': 0.8, 'label': 'Spearman ρ'},
    ax=ax,
)
ax.set_title('Spearman correlation heatmap (daily metrics + sentiment)')

out = REPORTS / 'nb_chartE_corr_heatmap.png'
plt.tight_layout()
plt.savefig(out, dpi=220)
plt.show()
print('Saved:', out)
```
